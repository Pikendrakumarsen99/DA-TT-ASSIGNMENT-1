```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



# Create health data using Pandas
data = pd.DataFrame({

    'Name' : ["pikendra","yash","harshit"],
    'Height': [165, 170, 160],
    'Weight': [86, 72, 80],
    'Steps': [8000, 6000, 10000],
    'Calories': [430, 180, 300],
    'Heart_Rate': [72, 85, 95]
})

# BMI calculation using NumPy on Pandas columns
data['BMI'] = data['Weight'] / (data['Height'] ** 2)

# Average values using NumPy
avg_steps = np.mean(data['Steps'])
avg_calories = np.mean(data['Calories'])
avg_hr = np.mean(data['Heart_Rate'])

#BMI BAR CHART
plt.figure()
plt.bar(data['Name'], data['BMI'])
plt.xlabel("Person Name")
plt.ylabel("BMI Value")
plt.title("BMI Analysis (Name Wise)")
plt.show()

# Steps Graph
plt.figure()
plt.plot(data['Name'], data['Steps'], marker='o')
plt.xlabel("Person Name")
plt.ylabel("Steps Count")
plt.title("Daily Steps Analysis (Name Wise)")
plt.show()

# Calories Graph
plt.figure()
plt.bar(data['Name'], data['Calories'])
plt.xlabel("Person Name")
plt.ylabel("Calories Burned")
plt.title("Calories Burned (Name Wise)")
plt.show()

# Heart Rate Graph
plt.figure()
plt.plot(data['Name'], data['Heart_Rate'], marker='o')
plt.xlabel("Person Name")
plt.ylabel("Heart Rate (bpm)")
plt.title("Heart Rate Monitoring (Name Wise)")
plt.show()

print("Average Steps:", avg_steps)
print("Average Calories:", avg_calories)
print("Average Heart Rate:", avg_hr)
```
